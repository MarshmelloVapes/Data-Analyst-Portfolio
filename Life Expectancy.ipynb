```python
# Import Libaries
from ast import increment_lineno
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in data
df = pd.read_csv('/Users/josh/Documents/Code/Data-Test-Life/life-expectancy.csv')
```

```python
# Find missing data, if any.
for col in df.columns:
    pct_missing = np.mean (df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for colums

df.dtypes
```

```python
# Print the data.
df
```

```python
# Get values for Canada only
df1 = df.loc[df["Entity"] == "Canada"]

# Get values for Zimbabwe only
df2 = df.loc[df["Entity"] == "Zimbabwe"]

df1
```

```python
# Compare year to life expectancy in Canada
# Scatter plot, year vs life expectancy.
plt.scatter(x=df1['Year'], y=df1['Life expectancy'])
plt.title('Year vs Life expectancy in Canada')
plt.xlabel('Year')
plt.ylabel('Life expectancy')
plt.show()
```

```python
# Seaborn, plot year vs life expectancy in Canada
sns.regplot(x='Year', y='Life expectancy', data=df1, scatter_kws={"color": "red"}, line_kws={"color": "black"})
plt.title('Year vs Life expectancy in Canada')
plt.show()
```

```python
# Mark correlation
#visualize info
correlation = df1.corr()

sns.heatmap(correlation, annot=True)
plt.title('Correlation in Canada')
plt.show()
```

```python
# Compare year to life expectancy in Zimbabwe.
# Scatter plot, year vs life expectancy.
plt.scatter(x=df2['Year'], y=df2['Life expectancy'])
plt.title('Year vs Life expectancy in Zimbabwe')
plt.xlabel('Year')
plt.ylabel('Life expectancy')
plt.show()
```

```python
# Seaborn, plot year vs life expectancy in Zimbabwe
sns.regplot(x='Year', y='Life expectancy', data=df2, scatter_kws={"color": "blue"}, line_kws={"color": "black"})
plt.title('Year vs Life expectancy in Zimbabwe')
plt.show()
```

```python
# Graph data from Canada and Zimbabwe to compare.
sns.regplot(x='Year', y='Life expectancy', data=df1, 
            scatter_kws={"color": "red"})
sns.regplot(x='Year', y='Life expectancy', data=df2, 
            scatter_kws={"color": "blue"})
plt.xlim(1945,2022)
plt.ylim(40, 90)
plt.title('Life Expectancy Canada vs Zimbabwe')
plt.show
```
